```python
import random
import time
import threading

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def parallel_merge_sort(arr, num_threads):
    if len(arr) <= 1:
        return arr

    if num_threads <= 1 or len(arr) < 1000:
        return merge_sort(arr)

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left_thread = threading.Thread(target=parallel_merge_sort, args=(left, num_threads // 2))
    right_thread = threading.Thread(target=parallel_merge_sort, args=(right, num_threads // 2))

    left_thread.start()
    right_thread.start()

    left_thread.join()
    right_thread.join()

    return merge(left, right)

def main():
    arr = [random.randint(1, 10000) for _ in range(10000)]

    # Merge Sort
    start_time = time.time()
    sorted_arr = merge_sort(arr.copy())
    merge_sort_time = time.time() - start_time

    # Multithreaded Merge Sort
    start_time = time.time()
    sorted_arr_parallel = parallel_merge_sort(arr.copy(), num_threads=4)
    parallel_merge_sort_time = time.time() - start_time

    print("Merge Sort Time:", merge_sort_time)
    print("Multithreaded Merge Sort Time:", parallel_merge_sort_time)

if __name__ == "__main__":
    main()
```
